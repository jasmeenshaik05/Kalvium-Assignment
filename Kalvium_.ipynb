```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Function to extract data, create DataFrame, and save to CSV
def extract_and_save(url, filename):
    # Fetching data from URL
    response = requests.get(url)
    print(response)

    # Parsing HTML with BeautifulSoup
    soup = BeautifulSoup(response.text, 'lxml')
    data_table = soup.find("table", class_="table")
    print(data_table)

    # Extracting table headers
    headers = []
    for th in data_table.find("thead").find_all("th"):
        headers.append(th.text.strip())
    print(headers)

    # Extracting table rows
    rows = []
    for tr in data_table.find_all("tr")[1:-1]:  # Skip header and footer rows
        cells = tr.find_all("td")
        row = []
        for cell in cells:
            row.append(cell.text.strip())
        rows.append(row)
    print(rows)

    # Creating a DataFrame
    df = pd.DataFrame(rows, columns=headers)
    print(df)

    # Saving DataFrame to CSV
    df.to_csv(filename, index=False)
    print(f"Saved {filename}")

# URLs and corresponding filenames
urls = [
    ("https://results.eci.gov.in/PcResultGenJune2024/index.htm", "election_results.csv"),
    ("https://results.eci.gov.in/AcResultGenJune2024/partywiseresult-S01.htm", "AndhraPradesh_election_results.csv"),
    ("https://results.eci.gov.in/AcResultGen2ndJune2024/partywiseresult-S02.htm", "ArunachalPradesh_election_results.csv"),
    ("https://results.eci.gov.in/AcResultGen2ndJune2024/partywiseresult-S21.htm", "Sikkim_election_results.csv"),
    ("https://results.eci.gov.in/AcResultGenJune2024/partywiseresult-S18.htm", "Odisha_election_results.csv")
]

# Execute extraction and save for each URL and filename pair
for url, filename in urls:
    extract_and_save(url, filename)
```
